```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import os

def open_folders():
    pic_parent_folder = pic_folder_entry.get()
    desc_parent_folder = desc_folder_entry.get()
    selected_date = date_entry.get()

    if not pic_parent_folder or not desc_parent_folder or not selected_date:
        return
    
    user_name = os.getlogin()  # Get the current user's username
    downloads_path = f"C:\\Users\\{user_name}\\Downloads"  # Construct the Downloads folder path for Windows

    
    pic_folder_path = os.path.join(downloads_path,  pic_parent_folder, selected_date[:7],  selected_date)
    print("Img folder full path : ", pic_folder_path)
    desc_folder_path = os.path.join(downloads_path, desc_parent_folder, selected_date[:7], selected_date)
    print("Desc folder full path : ", desc_folder_path)
    
    # Display images and descriptions from the specified date folder
    display_images_and_descriptions(pic_folder_path, desc_folder_path)

def display_images_and_descriptions(pic_folder, desc_folder):
    folder_listbox.delete(0, tk.END)  # Clear the listbox

    if not os.path.exists(pic_folder) or not os.path.exists(desc_folder):
        folder_listbox.insert(tk.END, "Folder not found.")
        return

    pic_files = os.listdir(pic_folder)
    desc_files = os.listdir(desc_folder)

    for pic_file, desc_file in zip(pic_files, desc_files):
        pic_path = os.path.join(pic_folder, pic_file)
        desc_path = os.path.join(desc_folder, desc_file)

        if os.path.isfile(pic_path) and os.path.isfile(desc_path):
            # Display the image
            image = Image.open(pic_path)
            image.thumbnail((200, 200))  # Resize for display
            img = ImageTk.PhotoImage(image)
            img_label = tk.Label(root, image=img)
            img_label.image = img  # To prevent image from being garbage collected
            img_label.pack()

            # Display details from the description file
            with open(desc_path, 'r') as desc_file:
                desc = desc_file.read()
                desc_label = tk.Label(root, text=desc)
                desc_label.pack()

# Create the main window
root = tk.Tk()
root.title("Display Images with Descriptions")

# Create labels and entries for pic and desc parent folders, and selected date
pic_folder_label = tk.Label(root, text="Enter Pic Parent Folder:")
pic_folder_label.pack(pady=5)
pic_folder_entry = tk.Entry(root)
pic_folder_entry.pack(pady=5)

desc_folder_label = tk.Label(root, text="Enter Desc Parent Folder:")
desc_folder_label.pack(pady=5)
desc_folder_entry = tk.Entry(root)
desc_folder_entry.pack(pady=5)

date_label = tk.Label(root, text="Enter Date (e.g., 2022-03-16):")
date_label.pack(pady=5)
date_entry = tk.Entry(root)
date_entry.pack(pady=5)

# Create a button to open folders and display images with descriptions
open_button = tk.Button(root, text="Open Folders", command=open_folders)
open_button.pack(pady=10)

# Create a listbox to display folder contents or messages
folder_listbox = tk.Listbox(root)
folder_listbox.pack(fill=tk.BOTH, expand=True)

# Start the Tkinter main loop
root.mainloop()
```
